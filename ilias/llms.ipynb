# Use LLMs to handle headlines

### Import the data

```python
import pandas as pd

# Read the seen and unseen data
train_seen = pd.read_parquet("../data/bars_seen_train.parquet")
train_unseen = pd.read_parquet("../data/bars_unseen_train.parquet")


headlines_seen = pd.read_parquet("../data/headlines_seen_train.parquet")
headlines_unseen = pd.read_parquet("../data/headlines_unseen_train.parquet")
```

### Call the LLM

```python
SESSION_ID = 701

# Convert filtered dataframe to string
subset = headlines_seen[headlines_seen["session"] == SESSION_ID][["headline", "bar_ix"]]
prompt_string = subset.to_string(index=False)

# Or better yet, format it as a list for the prompt:
formatted_list = "\n".join([f"Day {row.bar_ix}: {row.headline}" for _, row in subset.iterrows()])
formatted_list
```

```python
mid_cost = train_seen[train_seen["session"] == SESSION_ID]["close"].iloc[-1]
final_cost = train_unseen[train_unseen["session"] == SESSION_ID]["close"].iloc[-1]
print(final_cost - mid_cost)
```

```python
import anthropic
from dotenv import load_dotenv

load_dotenv()

client = anthropic.Anthropic()  # automatically reads ANTHROPIC_API_KEY

message = client.messages.create(
    model="claude-sonnet-4-6",
    max_tokens=1024,
    messages=[
        {"role": "user", "content": 
         "Explain recursion in simple terms."}
    ]
)

print(message.content[0].text)
```
